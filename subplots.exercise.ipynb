```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("company_data.csv")
df=df.head(15)
```

```python
df['Order Date']=df['Order Date'].str.extract(r'(^\d+)')
df['Order Date']=df['Order Date'].replace({'1':'january',
                                           '2':'february',
                                           '3':'march',
                                           '4':'april',
                                           '5':'may',
                                           '6':'june',
                                           '7':'july',
                                           '8':'august',
                                           '9':'september',
                                           '10':'october',
                                           '11':'november',
                                           '12':'december'})
df['Order Date']
```

```python
print(df.columns)
print()
```

```python
fig,axs=plt.subplots(3,1,figsize=(10,12))

#first plot
axs[0].plot(df['Order Date'].astype(str),df['Order Qty'],markerfacecolor='r')
axs[0].set(title='<<<QUANTITY ORDERED MONTHLY>>>',ylabel='__ORDERED QUANTITY__')
axs[0].grid(True)
plt.xticks(rotation=45)

#second plot

axs[1].bar(df['Order Date'].astype(str),df['Sales'],color='black')
axs[1].set(title='<<<SALES MADE PER MONTHLY>>>',ylabel='__SALES__')
axs[1].grid(True)
plt.xticks(rotation=35)

#THIRD PLOT

axs[2].bar(df['Order Date'].astype(str),df['Profit'],color='orange')
axs[2].set(title='<<<SALES MADE PER MONTHLY>>>',ylabel='**PROFIT**',xlabel='**MONTHS**')
axs[2].grid(True)
plt.xticks(rotation=25)

plt.show()           
```
