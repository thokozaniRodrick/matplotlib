```python
import matplotlib.pyplot as plt

data = pd.read_csv('temperature_data.csv')
data['Data Time'] = pd.to_datetime(data['Date Time'])
data=data.head(10)
plt.plot(data['Date Time'],data['T (degC)'],label='T (degC)',marker='*')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('temperature(*C)')
plt.title('Time_Temperature_Graph')
plt.show
```

```python
def Time_Temperature_Graph():
    data = pd.read_csv('temp_and_rain.csv')
    month_rain = data['rain']
    month = data['Month']
    colors=["blue","black","green","orange"]
    plt.xticks(rotation=45)
    bars=plt.bar(month,month_rain, color=colors)
    plt.xlabel('Monthly')
    plt.ylabel('rainfall(mm)')
    plt.title('Time_Temperature_Graph')
    plt.legend(bars,["january","february","march","april","may","june","july","august","september","october","november","december"])
    plt.show
Time_Temperature_Graph()

```
